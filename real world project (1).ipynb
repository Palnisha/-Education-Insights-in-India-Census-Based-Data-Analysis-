```python
DATAS.columns = DATAS.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
DATAS
```

```python
print("Total Records (Rows):", DATAS.shape[0])
print("Total Districts:", DATAS['area_name'].nunique())
print("Total Population:", DATAS['total_population_-_persons'].sum())
```

```python

## total male female studying
total_males_studying = DATAS['population_attending_educational_institutions_-_males'].sum()
total_females_studying = DATAS['population_attending_educational_institutions_-_females'].sum()

print("Total Males Studying:", total_males_studying)
print("Total Females Studying:", total_females_studying)
```

```python
### gender wis student studying
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
gender_data = [total_males_studying, total_females_studying]
labels = ['Males', 'Females']
plt.pie(gender_data, labels=labels, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Gender-wise Students Studying')
plt.show()

```

```python
# Rural vs Urban Population

# Rural Population
rural_population = DATAS[DATAS['total/_rural/_urban'] == 'Rural']['total_population_-_persons'].sum()

# Urban Population
urban_population = DATAS[DATAS['total/_rural/_urban'] == 'Urban']['total_population_-_persons'].sum()

plt.figure(figsize=(5,5))
area_data = [rural_population, urban_population]
labels = ['Rural', 'Urban']
plt.pie(area_data, labels=labels, autopct='%1.1f%%', colors=['orange', 'purple'])
plt.title('Rural vs Urban Population')
plt.show()
```

```python

#### total population male vs female
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_labels = ['Males','Females']
values = DATAS['total_population_-_males'].sum(),DATAS['total_population_-_females'].sum()
plt.bar(x_labels,values)
plt.ylabel('population')
plt.title('Total population Males vs Female')
plt.show()
```

```python
## total popolation vs people attending education

x_labels = ['total population','population attending educational institutions']
values = DATAS['total_population_-_persons'].sum(),DATAS['population_attending_educational_institutions_-_persons'].sum()
plt.bar(x_labels,values)
plt.ylabel('Count')
plt.title('Total population vs People attending education')
plt.show()
```

```python
##### age group vise education attendance
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import matplotlib.pyplot as plt
import seaborn as sns

# Group by Age-group and sum attending persons
age_group_data = DATAS.groupby("age-group")["population_attending_educational_institutions_-_persons"].sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=age_group_data, x="age-group", y="population_attending_educational_institutions_-_persons", palette="viridis")

plt.xticks(rotation=45)
plt.title("Educational Attendance Across Age Groups (Persons)")
plt.ylabel("Number of People Attending")
plt.xlabel("Age Group")
plt.tight_layout()
plt.show()
```

```python
### area name wise total population 

import matplotlib.pyplot as plt

# Create the line chart
x = DATAS['area_name']  
y = DATAS['total_population_-_persons'] 

plt.plot(x,y, marker='o', linestyle='-', color='b', label='Population')  
plt.title('Area Name wise Total Population')  
plt.xlabel('Area Name') 
plt.ylabel('Total Populations - Person')  
```

```python
## rural vs urban population
x_categories = {'Total': 1, 'Rural': 2, 'Urban': 3}  # Assign numeric values to categories
x = DATAS['total/_rural/_urban'].map(x_categories) 
y =DATAS['total_population_-_males']
z =DATAS['total_population_-_females']

# Create the scatter plot
plt.scatter(x, y, color='blue', label='Males')
plt.scatter(x, z, color='pink', label='Females')
plt.title('rural/urban wise Population: Males vs Females')
plt.xlabel('Rural/Urban')
plt.ylabel('Population')
plt.xticks(ticks=[1, 2, 3], labels=['Total', 'Rural', 'Urban'])
plt.legend()
plt.grid(True)  # Optional: Add a grid for better readability
plt.show()
```
